```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset
df_tracks = pd.read_csv('/Users/fanjaehyunmaenongdy/desktop/tracks.csv')
df_tracks.head()
```

```python
#Check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#Sort the dataframe by popularity in ascending order, fetch the first 10 rows.
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
#Top 10 most popular songs on Spotify
#With popularity > 90
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
#Here, I will change the current index to release date.
df_tracks.set_index('release_date',inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#Check artist at 18th row of Spotify Tracks Dataset.
df_tracks[['artists']].iloc[18]
```

```python
#Convert duration in ms to sec on Spotify Tracks Dataset.
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x:round(x/1000))
#Drop the duration_ms column.
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
#Plot the correlationship heatmap between varaibles.
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1,vmax=1, center=0,cmap="inferno",lw=0.75, linecolor="Black")
heatmap.set_title("Correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#From the above correlation heatmap,we can see the following major inference:
#Loudness and energy has a positive correlation of 0.8. So if the loudness of the track is high, that means the energy of the song is also high.
#Also, we see that valence and danceability has a positive correlation of 0.5.
#Popularity and acousticness has a negative correaltion of 0.4, if a track is higly popular then the acousticness of the track is low.
```

```python
#Considering 0.4% of the total dataset to create Regression plots.
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#Regression plot between loudness and energy.
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")
```

```python
#Regression plot between popularity and acousticness.
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="r").set(title="Popularity vs Acousticness Correlation")
```

```python
#Let's find the total number of songs each year since 1992.
#First, creating new columns in tracks dataset (Year, Release Date).
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
#The total number of songs each year since 1922.
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
#Create a bar plot
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year VS Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year VS Duration")
plt.xticks(rotation=60)
```

```python
#As you can see, the duration of songs from 1922 to 2010 were increasing. However, after 2010 the trend was declining.
```
